```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("/content/Individual_symptom.csv")

# Transpose the dataframe to have the labels as columns
df_transposed = df.transpose()

# Set the column names to the first row values
df_transposed.columns = df_transposed.iloc[0]

# Remove the first row (since it's now the column names)
df_transposed = df_transposed.iloc[1:]

# Convert values to numeric
df_transposed = df_transposed.apply(pd.to_numeric)

# Plot
plt.figure(figsize=(10, 6))
sns.heatmap(df_transposed, cmap="viridis", annot=True, fmt=".0f", linewidths=.5, cbar=False)
plt.title("Label Distribution")
plt.xlabel("Labels")
plt.ylabel("Symptoms")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
